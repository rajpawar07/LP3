```python
x=2 # starting point
learningRate = 0.01
precision = 0.000001 #when to stop algo
prev_step_size = 1
max_iter = 10000
iters = 0
gradient_function = lambda x: (x+3)**2
```

```python
import matplotlib.pyplot as plt
gd =[]
```

```python
while precision < prev_step_size and iters < max_iter:
    prev = x
    x = x-learningRate*gradient_function(prev)
    prev_step_size = abs(x - prev)
    iters += 1
    print('Iteration : ',iters,'Values : ',x)
    gd.append(x)
```

```python
print('Local Minima : ',x)
```

```python
plt.plot(gd)
```
